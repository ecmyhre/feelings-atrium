<div class="alert alert-success">NumPy's math functions and other operations can be applied to Series without losing the data structure.  </div>

<div class="alert alert-success"> A Series can be thought of as an ordered key-value store. In fact, we can create one from a dict:</div>

```python
theliars_dict = {'spencer': 632,
                'emily': 1638,
                'hanna': 569,
                'arya': 115}
```

```python
theliars_dict
```

<div class="alert alert-success"> Notice that the Series is created in key-sorted order.</div>

<div class="alert alert-success">If we pass a custom index to Series, it will select the corresponding values from the dict, and treat indices without corrsponding values as missing. Pandas uses the NaN (not a number) type for missing values.  </div>

```python
A = pd.Series(theliars_dict, index=['spencer',
                                   'mona',
                                   'ezra',
                                   'arya'])

A
```

<div class="alert alert-danger"> Critically, the labels are used to align data when used in operations with other Series objects:

```python
prettylittleliars + A
```

<div class="alert alert-success"> Contrast this with NumPy arrays, where arrays of the same length will combine values element-wise; adding Series combined values with the same label in the resulting series. Notice also that the missing values were propogated by addition.  </div>

### cor and cov

We can also calculate summary statistics *across* multiple columns, for example, correlation and covariance.

$$cov(x,y) = \sum_i (x_i - \bar{x})(y_i - \bar{y})$$

```python
baseball.hr.cov(baseball.X2b)
```

$$corr(x,y) = \frac{cov(x,y)}{(n-1)s_x s_y} = \frac{\sum_i (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_i (x_i - \bar{x})^2 \sum_i (y_i - \bar{y})^2}}$$

```python
baseball.hr.corr(baseball.X2b)
```

```python
baseball.ab.corr(baseball.h)
```

```python
baseball.corr()
```

<div class="alert alert-success">If we have a DataFrame with a hierarchical index (or indices), summary statistics can be applied with respect to any of the index levels:</div>
